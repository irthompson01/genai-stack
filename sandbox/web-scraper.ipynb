## Web Scraping for RAG App

```python
# import libraries for webscarping and data parsing

import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

# Path: web-scraper.ipynb
```

```python
# Given a url, return all sub pages of the base url
# Example: www.example.com/1, www.example.com/2, www.example.com/3 from www.example.com

def get_sub_pages(url):
    # Get the base url
    # base_url = url.split('/')[0] + '//' + url.split('/')[2]
    base_url = url
    print(base_url)
    
    # Get the page
    page = requests.get(url)
    
    # Parse the page
    soup = BeautifulSoup(page.content, 'html.parser')
    
    # Get all the links
    links = soup.find_all('a')
    
    # Get the sub pages
    sub_pages = []
    for link in links:
        # print(link)
        # print(type(link))
        href = link.get('href')
        
        if href is not None and href[:5].isdigit():
            sub_pages.append(base_url + link.get('href'))
            
    return sub_pages

url = "https://docs.data.world/en/"

results = get_sub_pages(url)

print(len(results))

print(results)
```
